# What are pokmon?

Pokmon are creatures of all shapes and sizes who live in the wild or alongside humans. For the most part, Pokmon do not speak except to utter their names. Pokmon are raised and commanded by their owners (called Trainers). During their adventures, Pokmon grow and become more experienced and even, on occasion, evolve into stronger Pokmon. There are currently more than 700 creatures that inhabit the Pokmon universe.

They can be of different types: fire, water, electric, grass, ghost and so on. They can also be strong or weak, as well as slow or fast, and legendary or not. 



As we have several pokmons, we can analyze the large volume of data that they are in, to determine characteristics they have, such as: which type of pokemon has the highest average, which is the fastest, which has the highest defense, which one has the best attack? Come and join this adventure...

# Cleaning data

```python
#Renaming columns
pokemon.rename(columns={'Type 1': 'Type_1',
                        'Type 2': 'Type_2',
                        'Sp. Atk': 'Sp_Atk',
                        'Sp. Def': 'Sp_Def'},
                        inplace=True)
pokemon.columns
```

```python
#COnverting type of the varialbes
pokemon['Name']=pokemon['Name'].astype("string")
pokemon['Type_1']=pokemon['Type_1'].astype("category")
pokemon['Type_2']=pokemon["Type_2"].astype("category")
pokemon['Generation']=pokemon["Generation"].astype("category")
```

```python
pokemon.dtypes
```

```python
#Filling the missing data
pokemon['Type_2'] = np.where(pokemon['Type_2'].isnull(), pokemon['Type_1'],pokemon['Type_2'])
```

```python
pokemon.isnull().any()
```

# Grouping by

```python
#Grouping Type_1's values
grouped_main = pokemon.groupby("Type_1")
```

```python
type(grouped_main)
```

# Visualizing and analyzing

```python
#The type of pokemon which has the highest average HP
grouped_main['HP'].mean().sort_values(ascending=False)
```

```python
#Visualize the results
fig, ax = plt.subplots(1, 1)
group_obj = grouped_main['HP'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=pokemon, x='Type_1', y='HP', palette='bright', ax=ax, ci=None,order=group_obj['Type_1'])
xlabels = ax.get_xticklabels()
ax.set_xticklabels(rotation=90, labels=xlabels)
ax.set_xlabel('Pokemon Type', fontsize=14),
ax.set_ylabel('HP', fontsize=14);
```

```python
#The fastest type of pokemon
grouped_main['Speed'].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(1, 1)
group_obj = grouped_main['Speed'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=pokemon,x='Type_1',y='Speed', palette='Paired',ax=ax, ci=None,order=group_obj['Type_1'])
xlabels = ax.get_xticklabels()
ax.set_xticklabels(rotation=90, labels=xlabels)
ax.set_xlabel('Pokemon Type', fontsize=14),
ax.set_ylabel('Speed', fontsize=14);
```

```python
#The type of pokemon which has the best attack 
grouped_main["Attack"].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(1, 1)
group_obj = grouped_main['Attack'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=pokemon,x='Type_1',y='Attack', palette='Set1',ax=ax, ci=None,order=group_obj['Type_1'])
xlabels = ax.get_xticklabels()
ax.set_xticklabels(rotation=90, labels=xlabels)
ax.set_xlabel('Pokemon Type', fontsize=14),
ax.set_ylabel('Attack', fontsize=14);
```

```python
#The type of pokemon which has the strongest defense
grouped_main['Defense'].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(1, 1)
group_obj = grouped_main['Defense'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=pokemon,x='Type_1',y='Defense', palette='tab10',ax=ax, ci=None,order=group_obj['Type_1'])
xlabels = ax.get_xticklabels()
ax.set_xticklabels(rotation=90, labels=xlabels)
ax.set_xlabel('Pokemon Type', fontsize=14),
ax.set_ylabel('Defense', fontsize=14);
```

```python
#Type of pokemon which has the strongest total strength on average
grouped_main['Total'].mean().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(1, 1)
group_obj = grouped_main['Total'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=pokemon,x='Type_1',y='Total', palette='pastel',ax=ax, ci=None,order=group_obj['Type_1'])
xlabels = ax.get_xticklabels()
ax.set_xticklabels(rotation=90, labels=xlabels)
ax.set_xlabel('Pokemon Type', fontsize=14),
ax.set_ylabel('Total', fontsize=14);
```

```python
#Grouping by Total, Generation and Legendary
sns.catplot(x='Total', y='Generation',hue='Legendary', kind='bar',palette='Set1',data=pokemon,ci=None)
```

```python
#Grouping by Total, Generation and Legendary in different  graphs
sns.catplot(y='Total', x='Generation',col='Legendary', kind='bar',palette='Set1',data=pokemon, ci=None)
```
